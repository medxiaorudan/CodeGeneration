# **LangChain**

[LangChain](https://github.com/langchain-ai/langchain) is a framework for developing applications powered by language models. It enables applications that:

- **Context-aware:** Connect a language model to sources of context (prompt instructions, few shot examples, content to ground its response in, etc.)
- **Reason:** Rely on a language model to reason (about how to answer based on provided context, what actions to take, etc.)

This framework consists of several parts.

### LangChain Libraries

The LangChain Libraries include Python and JavaScript libraries. They contain interfaces and integrations for a myriad of components, a basic runtime for combining these components into chains and agents, and off-the-shelf implementations of chains and agents.

### LangChain Templates

LangChain Templates are a collection of easily deployable reference architectures for a wide variety of tasks.

### LangServe

LangServe is a library for deploying LangChain chains as a REST API.

### LangSmith

LangSmith is a developer platform that lets you debug, test, evaluate, and monitor chains built on any LLM (Large Language Model) framework. It seamlessly integrates with LangChain.

```python
import langchain
import os
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain_experimental.smart_llm import SmartLLMChain
```

```python
with open("/content/drive/MyDrive/Key.txt", "r") as f:
    key = f.read()

os.environ["OPENAI_API_KEY"] = key
```

```python
question="Generate a C++ program that accepts numeric input from the user and maintains a record of previous user inputs with timestamps. Ensure the program sorts the user inputs in ascending order based on the provided numeric input. Enhance the program to display timestamps along with the sorted user inputs."
```

## **Traditional output method with gpt-3.5-turbo**

```python
prompt=PromptTemplate.from_template(question)
llm=ChatOpenAI(temperature=0.3, model_name="gpt-3.5-turbo")

from langchain.chains import LLMChain
default_chain=LLMChain(llm=llm, prompt=prompt, verbose=True)
default_chain.run(prompt=question)
```

## **Use SmartChain method with gpt-3.5-turbo**

```python
chain=SmartLLMChain(llm=llm, prompt=prompt, n_ideas=3, verbose=True)
chain.run({})
```

## **Use SmartChain method with gpt-3.5-turbo and GPT-4**

```python
chain2=SmartLLMChain(
    ideation_llm=ChatOpenAI(temperature=0.9, model_name="gpt-3.5-turbo"),
    llm=ChatOpenAI(temperature=0, model_name="gpt-4"),
    prompt=prompt,
    n_ideas=3,
    verbose=True
)

chain2.run({})
```

## **Use SmartChain method with GPT-4**

```python
chain3=SmartLLMChain(
    ideation_llm=ChatOpenAI(temperature=0.9, model_name="gpt-4"),
    critique_llm=ChatOpenAI(temperature=0.5, model_name="gpt-4"),
    resolver_llm=ChatOpenAI(temperature=0, model_name="gpt-4"),
    prompt=prompt,
    n_ideas=3,
    verbose=True
)

chain3.run({})
```
