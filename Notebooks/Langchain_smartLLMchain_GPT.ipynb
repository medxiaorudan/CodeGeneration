# **LangChain**

[LangChain](https://github.com/langchain-ai/langchain) is a framework for developing applications powered by language models. It enables applications that:

- **Context-aware:** Connect a language model to sources of context (prompt instructions, few shot examples, content to ground its response in, etc.)
- **Reason:** Rely on a language model to reason (about how to answer based on provided context, what actions to take, etc.)

This framework consists of several parts.

### LangChain Libraries

The LangChain Libraries include Python and JavaScript libraries. They contain interfaces and integrations for a myriad of components, a basic runtime for combining these components into chains and agents, and off-the-shelf implementations of chains and agents.

### LangChain Templates

LangChain Templates are a collection of easily deployable reference architectures for a wide variety of tasks.

### LangServe

LangServe is a library for deploying LangChain chains as a REST API.

### LangSmith

LangSmith is a developer platform that lets you debug, test, evaluate, and monitor chains built on any LLM (Large Language Model) framework. It seamlessly integrates with LangChain.

```python
question1="Generate a C++ program that accepts numeric input from the user and maintains a record of previous user inputs with timestamps. Ensure the program sorts the user inputs in ascending order based on the provided numeric input. Enhance the program to display timestamps along with the sorted user inputs."

question2="Generate a C++ program that accepts numeric input from the user and maintains a record of previous user inputs with timestamps. \
Ensure the program sorts the user inputs in ascending order based on the provided numeric input. \
Include error handling to handle non-numeric inputs and provide a message to the user when exiting the program. \
Use a suitable data structure (e.g., std::vector or std::list) to efficiently store and manage the user input records. \
Implement a function to display the sorted user input records with timestamps. \
Enhance the program to display timestamps along with the sorted user inputs. \
Use the C++ standard library (e.g., std::chrono::system_clock) for handling timestamps. \
Ensure that the timestamp is updated each time a new user input is added. \
Provide proper formatting for the timestamp in the output."
```

## **Traditional output method with gpt-3.5-turbo**

```python
prompt1=PromptTemplate.from_template(question1)
llm=ChatOpenAI(temperature=0.3, model_name="gpt-3.5-turbo")

from langchain.chains import LLMChain
default_chain=LLMChain(llm=llm, prompt=prompt1, verbose=True)
default_chain.run(prompt=question1)
```

```python
prompt2=PromptTemplate.from_template(question2)
llm=ChatOpenAI(temperature=0.3, model_name="gpt-3.5-turbo")

from langchain.chains import LLMChain
default_chain=LLMChain(llm=llm, prompt=prompt2, verbose=True)
default_chain.run(prompt=question2)
```

## **Use SmartChain method with gpt-3.5-turbo**

```python
chain=SmartLLMChain(llm=llm, prompt=prompt1, n_ideas=3, verbose=True)
chain.run({})
```

```python
chain=SmartLLMChain(llm=llm, prompt=prompt2, n_ideas=3, verbose=True)
chain.run({})
```
